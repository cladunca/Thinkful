```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('pokemon.csv')
```

```python
df.head()
```

```python
types = df['Type 1']
totals = df['Total']
plt.hist(totals, bins=50, color = 'violet')
plt.ylabel('Total Pokemon')
plt.xlabel('Total Stats')
plt.show()
```

```python
plt.boxplot(totals)
plt.title('Boxplot')
plt.ylabel('Total Stats')
plt.show()
```

```python
plt.plot(totals)
plt.show()
```

```python
p = sns.lmplot(y='HP',
               x='Total',
               data=df,
               fit_reg=False,
               scatter_kws={'alpha':0.5})
plt.title('Scatterplot: Pokemon HP vs Total Stats')
plt.show()

p = sns.lmplot(y='HP',
               x='Total',
               data=df,
               fit_reg=True,
               scatter_kws={'alpha':0.5})
plt.title('Scatterplot with Regression Line: Pokemon HP vs Total Stats')
plt.show()
```

```python
plt.scatter(x=df['HP'], y=totals)
plt.xlabel('HP')
plt.ylabel('Total Stats')
plt.show()
# Is there a correlation between HP and total stats?
```

```python
p = sns.lmplot(y='HP', 
               x='Total',
               data=df,
               scatter_kws={'alpha':0.4}, 
               palette='pastel',
               ) 
p.set_ylabels("Tips")
p.set_xlabels("Total Bill")
plt.title('Scatterplot Lowess: Meal Tips by Total Bill, split by time and gender')
plt.show()
```

```python
# compare type1 and attack value

attack = df['Attack']

plt.subplots(figsize=(20,15))

sns.set(style="darkgrid")

g = sns.barplot(x=types, y=attack, data=df)
g.set_ylabel("Attack")
g.set_xlabel("Pokemon Main Type")
plt.title('Barplot:Pokemon Attack by Primary Type')
plt.show()
```

```python
sns.set(style="darkgrid")
plt.subplots(figsize=(20,15))


g = sns.swarmplot(x=types, y=attack, data=df)
g.set_ylabel("Attack")
g.set_xlabel("Pokemon Main Type")
plt.title('Barplot:Pokemon Attack by Primary Type')
plt.show()
```

```python
sns.set(style="darkgrid")
plt.subplots(figsize=(20,15))


g = sns.stripplot(x=types, y=attack, data=df)
g.set_ylabel("Attack")
g.set_xlabel("Pokemon Main Type")
plt.title('Barplot:Pokemon Attack by Primary Type')
plt.show()
```
