```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
binomial = np.random.binomial(50, .3, 100)
```

```python
plt.hist(binomial, color='purple')
plt.axvline(binomial.mean(), color='green', linestyle='solid', linewidth=2)
plt.axvline(binomial.mean()+binomial.std(), color='green', linestyle='dashed', linewidth=2)
plt.axvline(binomial.mean()-binomial.std(), color='green', linestyle='dashed', linewidth=2)

plt.show()
```

```python
v1 = np.random.normal(5,.5, 100)
v2 = np.random.normal(10,1, 100)
v3 = v1+v2

mean = np.mean(v3)
std = np.std(v3)
plt.hist(v3)
plt.axvline(mean,color='green')
plt.axvline(mean+std, color='green', linestyle='dashed')
plt.axvline(mean-std, color='green', linestyle='dashed')
plt.show()
```
