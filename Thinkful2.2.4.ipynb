```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Table_13.02_GROSS_DOMESTIC_PRODUCT__TOTAL_AND_PER_CAPITA_AND_RESIDENT_POPULATION_1963_TO_2014.csv')
```

```python
print(df)
```

```python
plt.plot(df['GDP (current), $million'], color='magenta')
plt.ylabel('GDP ($million)')
plt.ylim([0, 80000])
plt.xlabel('Years since 1963')
plt.title('GDP Growth in Hawaii 1963-2014')
plt.show()

# This chart shows GDP growth in Hawaii since 1963. I tried casting the year to the x axis but was
# unsuccessful in doing so. While the Aloha State's economy has seen significant growth since
# 1963, it had a couple of slumps, one being in the '90s and then again in the recession
# around 2008.
```

```python

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(df['GDP (current), $million'], color='magenta')
plt.ylabel('GDP ($million)')
plt.ylim([0, 80000])
plt.xlabel('Years since 1963')
plt.title('GDP Growth in Hawaii 1963-2014')


plt.subplot(1, 2, 2)
plt.plot(df['Resident population'], color='cyan')
plt.ylabel('Population')
plt.xlabel('Years since 1963')
plt.title('Population Growth in Hawaii 1063-2014')
plt.tight_layout()

plt.show()

# These two plots show side-by-side the GDP growth and population growth of
# Hawaii 1963-2014. While Hawaii's population doubled in 50 years, its
# economy increased at a much larger rate, showing that there are more
# variables involved in the growth of an economy beyond the number of people
# contributing to it.
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(y=df['GDP (current), $million'], x=df['Resident population'], color='green')
plt.xlabel('Population')
plt.ylabel('GDP ($million)')
plt.title('Population Growth vs. GDP growth in Hawaii 1963-2014')

plt.show()

# This chart shows similar information to what is in the charts above. However,
# instead of showing GDP and population growth each as a function of time, here
# we have a scatter plot showing GDP growth as a function of population growth.
# Hawaii's economy has grown as its population has, though the relation is not
# one-to-one
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['Resident population'], bins=20, color='purple')
plt.title('Population of Hawaii')
plt.xlabel('Residents')
plt.show()

# This data set doesn't readily lend itself towards plots such as histograms.
# What I've gathered here is that the shorter bins represent faster periods of
# growth since there are fewer data points in those bins. In particular, I
# noticed that when Hawaii hit roughly 1.2 million people (the tallest bin), 
# its population stagnated a bit before resuming growth towards 1.4 million people. 
# This information is also reflected in the cyan-colored line chart above, where 
# you see a bit of a slump around the 1.2 million mark.
```
